##  **Dataset 1:**

- The data contains recored values of **concentration of various pollutants** in various **cities of india** from a time span of **2015 to 2020.** The observations are recorded **daily.**

- The data also contains the Air Quality Index (AQI) value for each day along the Air Quality class (**Good, Satisfactory, Moderate, Poor, Very Poor, and Severe.**)

```python
# Dataset 1
path = 'https://gitlab.com/academics3/insaid-data/-/raw/main/Career%20Wednesday%20Repo/17th%20August%202022%3A%20Data%20Manipulation%20Assessment/Air_Quality_index.csv'

# First Look
df = pd.read_csv(path, index_col=['Date'])
df.head(7)
```

## **Question 1:**

- **Industrial waste** is the combination of SO2, O3, Benzene, Toluene, and Xylene.

- **Vehicle waste** is the combination of  PM2.5, PM10, NO2, NH3, and CO.

- And **Total waste** is the **sum** of Industrial waste and vehicle waste.

- Using this information format the dataframe and find the **average total waste each city has produced.**

- Finally identify the **top 7 Most Polluted cities** based on Total waste from the obtained results.

```python
# Explore

# Creating the combined dataframe of Indutrial and Vehicle Waste
df['Industrial_Waste'] = df[['SO2', 'O3', 'Benzene', 'Toluene', 'Xylene']].sum(axis=1) #Adding all Industry related waste
df['Vehicle_Waste'] = df[['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO']].sum(axis=1) #Adding all vehicle elated waste
df['Total_Waste'] = df[['Industrial_Waste', 'Vehicle_Waste']].sum(axis=1) #Total Pollutants in the air

# Lets take a look at combined dataframe
df.head(7)
```

```python
# Lets calculate the average for each city
df_tmp = df.groupby('City').mean()
df_tmp.head()
```

```python
# Sorting based on Total Waste
df_tmp = df_tmp.sort_values('Total_Waste', ascending=False)
df_tmp.head(7)
```

```python
# Extracting Top 7 Cities
df_tmp = df_tmp.reset_index()[['City', 'Total_Waste']]
df_tmp.loc[:6]
```

```python
# Answer
Top_7_polluted_cities = ['Delhi', 'Patna', 'Ahmedabad', 'Gurugram', 'Luckhnow', 'Mumbai', 'Kolkata']

# Checking your Answer (Don't Change this Code)
CheckingAnswer('Top7', Top_7_polluted_cities) # Delhi, Patna, Ah, Guru, Luck, Eu, Mumbai
```
