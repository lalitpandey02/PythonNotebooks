```python
def create_model():
  model=Sequential()
  model.add(Dense(12,input_shape=(8,),activation='relu'))
  model.add(Dense(1,activation='sigmoid'))
  model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
  return model
```

```python
data=pd.read_csv('diabetes.csv')
```

```python
X=data.iloc[:,0:8]
Y=data.iloc[:,8]
```

```python
model=KerasClassifier(model=create_model,verbose=1)
```

```python
def create_model(optimizer='adam'):
  model=Sequential()
  model.add(Dense(12,input_shape=(8,),activation='relu'))
  model.add(Dense(1,activation='sigmoid'))
  model.compile(loss='binary_crossentropy',optimizer=optimizer,metrics=['accuracy'])
  return model
```

```python
model=KerasClassifier(model=create_model,epochs=100,batch_size=10,verbose=1)
```

```python
optimizer=['RMSprop','Adagrad','Adam']
```

```python
param_grid=dict(model__optimizer=optimizer)
```

```python
grid=GridSearchCV(estimator=model,param_grid=param_grid,n_jobs=-1,cv=2)
```

```python
grid_result=grid.fit(X,Y)
```

```python
means=grid_result.cv_results_['mean_test_score']
stds=grid_result.cv_results_['std_test_score']
params=grid_result.cv_results_['params']
```

```python
for mean, std, param in zip(means,stds,params):
  print(mean,std,param)
```

```python
def create_model(optimizer='adam',neurons=5,dropout_rate=0.2):
  model=Sequential()
  model.add(Dense(10,input_shape=(8,),activation='relu'))
  model.add(Dense(neurons,activation='relu'))
  model.add(Dropout(dropout_rate))
  model.add(Dense(1,activation='sigmoid'))
  model.compile(loss='binary_crossentropy',optimizer=optimizer,metrics=['accuracy'])
  return model
```

```python
model=KerasClassifier(model=create_model,epochs=100,batch_size=10,verbose=1)
```

```python
optimizer=['RMSprop','Adagrad','Adam']
dropout_rate=[0.1,0.2,0.3]
neurons=[10,15,20]
param_grid=dict(model__neurons=neurons,model__dropout_rate=dropout_rate,model__optimizer=optimizer)
```

```python
grid=GridSearchCV(estimator=model,param_grid=param_grid)
```

```python
grid_result=grid.fit(X,Y)
```
